```python
import matplotlib.pyplot as plt

# Data F1-score terbaik dari tabelmu
datasets = ["Original", "U-10k", "U-20k", "U-30k"]
f1_scores = [67.75, 74.96, 76.07, 77.52]

# Warna berbeda untuk setiap bar
colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2']

plt.figure(figsize=(8,5))

bars = plt.bar(datasets, f1_scores, color=colors)

# Tambahkan nilai pada setiap bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height}%",
        ha='center',
        va='bottom'
    )

plt.ylabel("F1-score (%)")
plt.title("Perbandingan Performa Model IndoBERT\n(F1-score Terbaik Tiap Dataset)")
plt.tight_layout()

plt.savefig("perbandingan_f1_indobert.png", dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dataset dan F1-score terbaik CNN-BiLSTM
datasets = ["Original", "U-10k", "U-20k", "U-30k"]
f1_scores = [67.85, 68.29, 70.58, 71.72]

# Warna untuk setiap bar
colors = ['#4E79A7', '#F28E2B', '#E15759', '#76B7B2']

plt.figure(figsize=(8,5))

bars = plt.bar(datasets, f1_scores, color=colors)

# Tambahkan angka di atas bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f"{height}%",
        ha='center',
        va='bottom'
    )

plt.ylabel("F1-score (%)")
plt.title("Perbandingan Performa Model CNN-BiLSTM\n(F1-score Terbaik Tiap Dataset)")
plt.tight_layout()

plt.savefig("perbandingan_f1_cnn_bilstm_colored.png", dpi=300)
plt.show()
```
