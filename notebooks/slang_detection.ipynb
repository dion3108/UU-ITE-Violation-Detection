```python
import re
import pandas as pd

# === 1. Load dataset ===
df = pd.read_csv("Data/gabungan_komentar_combined_filled.csv")


# === 2. Case Folding + Tokenisasi + Hapus simbol/emotikon/mention ===
def preprocess(text):
    text = text.lower()                          # case folding
    text = re.sub(r'@[A-Za-z0-9_]+', ' ', text)  # hapus mention
    text = re.sub(r'[^a-z0-9\s]', ' ', text)     # hapus simbol & emotikon
    tokens = re.findall(r'\b\w+\b', text)        # tokenisasi
    return tokens

df['tokens'] = df['Komentar'].astype(str).apply(preprocess)

# === 3. Remove Duplicate Komentar ===
df = df.drop_duplicates(subset=["Komentar"]).reset_index(drop=True)

# === 4. Remove Duplicate Token di dalam satu komentar ===
df['tokens_unique'] = df['tokens'].apply(lambda x: list(set(x)))

# === 5. Daftar Token Unik Global ===
all_tokens = [tok for row in df['tokens_unique'] for tok in row]
unique_tokens = list(set(all_tokens))

print("Jumlah komentar unik:", len(df))
print("Jumlah total token:", len(all_tokens))
print("Jumlah token unik global:", len(unique_tokens))
print("Contoh token unik:", unique_tokens[:30])
```

```python
# === 1. Load kamus KBBI dari file txt ===
with open("Data/kbbi.txt", "r", encoding="utf-8") as f:
    kbbi_words = {line.strip().lower() for line in f if line.strip()}

# === 2. Filter hanya kata tunggal (tanpa spasi/strip) ===
kbbi_single_words = {w for w in kbbi_words if " " not in w and "-" not in w}

print("Jumlah kata di KBBI:", len(kbbi_single_words))

# === 3. Cari kandidat slang ===
candidate_slang = [word for word in unique_tokens if word not in kbbi_single_words]

print("Jumlah kandidat slang:", len(candidate_slang))
print("Contoh kandidat slang:", candidate_slang[:50])
```

```python
# IMPORTANT: Make sure to use the "slang-uuite" environment kernel
# In VS Code: Ctrl+Shift+P → "Python: Select Interpreter" → Choose "slang-uuite"

import gensim

# Load FastText pre-trained model (.vec format)
fasttext_model = gensim.models.KeyedVectors.load_word2vec_format(
    "Data/cc.id.300.vec/cc.id.300.vec", binary=False
)

print("FastText model loaded successfully using gensim!")
```

```python
# Fungsi untuk Cari Kata Normal Terdekat

def most_similar_baku(word, kbbi_words, model, topn=1):
    if word not in model:
        return None, 0.0
    sims = []
    for w in kbbi_words:
        if w in model:
            score = model.similarity(word, w)
            sims.append((w, score))
    if not sims:
        return None, 0.0
    sims = sorted(sims, key=lambda x: -x[1])
    return sims[0]  # kata paling mirip
```

```python
# Normalisasi Kandidat Slang
slang_dict = {}
threshold = 0.05  # bisa diubah sesuai eksperimen

for slang in candidate_slang[:500]:  # batasi dulu 500 slang pertama untuk testing
    normal, score = most_similar_baku(slang, kbbi_single_words, fasttext_model)
    if score >= threshold:
        slang_dict[slang] = (normal, score)

print("Jumlah slang yang berhasil dinormalisasi:", len(slang_dict))
print("Contoh mapping slang → normal:")
for k, v in list(slang_dict.items())[:20]:
    print(f"{k} → {v[0]} (similarity={v[1]:.3f})")
```

```python
# Simpan Kamus Slang
import pandas as pd

slang_df = pd.DataFrame([
    {"slang": slang, "normal": normal, "similarity": score}
    for slang, (normal, score) in slang_dict.items()
])
slang_df.to_csv("slang_dictionary.csv", index=False)
```
