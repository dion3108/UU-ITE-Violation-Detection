```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load CSV
df = pd.read_csv("no_duplicated_undersampled_52000_nonbully.csv")

# Clean label values
df['final_label'] = df['final_label'].replace({'-': '0', ' ': '0', '0\\': '0'})

# Convert to integer
df['final_label'] = df['final_label'].astype(int)

# Separate majority and minority classes
df_0 = df[df['final_label'] == 0]
df_1 = df[df['final_label'] == 1]

# Choose how many samples you want from class 0 (e.g., 42000, 32000, 22000)
target_count = 42000  # Change this value to 32000 or 22000 as needed

# Undersample class 0
df_0_undersampled = df_0.sample(n=target_count, random_state=42)

# Combine with all class 1
df_balanced = pd.concat([df_0_undersampled, df_1], ignore_index=True)

# Shuffle
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# First split: 70% train, 30% temp
train_df, temp_df = train_test_split(df_balanced, test_size=0.3, stratify=df_balanced['final_label'], random_state=42)

# Second split: 15% val, 15% test
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['final_label'], random_state=42)

# Check distribution
print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
print(train_df['final_label'].value_counts())
print(val_df['final_label'].value_counts())
print(test_df['final_label'].value_counts())

# Save
train_df.to_csv("40k/train.csv", index=False)
val_df.to_csv("40k/val.csv", index=False)
test_df.to_csv("40k/test.csv", index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load CSV
df = pd.read_csv("no_duplicated_undersampled_52000_nonbully.csv")

# Clean 'final_label' column
df['final_label'] = df['final_label'].replace({'-': '0', ' ': '0', '0\\': '0'})
df['final_label'] = df['final_label'].astype(int)

# Drop how many from final_label == 0
drop_count = 30000  # Change to 20000 or 30000 if needed

# Split the data
df_0 = df[df['final_label'] == 0]
df_1 = df[df['final_label'] == 1]

# Drop specified number of class 0 rows
df_0_dropped = df_0.sample(n=len(df_0) - drop_count, random_state=42)

# Combine class 0 and class 1
df_balanced = pd.concat([df_0_dropped, df_1], ignore_index=True)
df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

# Split into train (70%), val (15%), test (15%)
train_df, temp_df = train_test_split(df_balanced, test_size=0.3, stratify=df_balanced['final_label'], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['final_label'], random_state=42)


# Check distribution
print(f"Train: {len(train_df)}, Val: {len(val_df)}, Test: {len(test_df)}")
print(train_df['final_label'].value_counts())
print(val_df['final_label'].value_counts())
print(test_df['final_label'].value_counts())

# Save to CSV
train_df.to_csv("20k/train.csv", index=False)
val_df.to_csv("20k/val.csv", index=False)
test_df.to_csv("20k/test.csv", index=False)
```
